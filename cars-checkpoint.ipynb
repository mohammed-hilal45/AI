```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.metrics import mean_absolute_error
```

```python
data = pd.read_csv("cars-data.csv")
```

```python
print("\nأول 5 صفوف من البيانات:")
print(data.head())
```

```python
data = data.drop(['Model'], axis=1)
```

```python
data.dropna(inplace=True)
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data['Price'], bins=50, kde=True, color='blue')
plt.title('توزيع الأسعار قبل التحويل اللوغاريتمي')
plt.show()
```

```python
for column in ['Price', 'Mileage', 'EngineV']:
    q = data[column].quantile(0.99)
    data = data[data[column] < q]
```

```python
q = data['Year'].quantile(0.01)
data = data[data['Year'] > q]
```

```python
data['log_price'] = np.log(data['Price'])
data = data.drop(['Price'], axis=1)
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data['log_price'], bins=50, kde=True, color='green')
plt.title('توزيع الأسعار بعد التحويل اللوغاريتمي')
plt.show()
```

```python
features = data[['Mileage', 'EngineV', 'Year']]
vif = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]
print("\nمؤشر تعدد التداخل بين المتغيرات (VIF):", vif)
```

```python
data = data.drop(['Year'], axis=1)
```

```python
x = data.drop(['log_price'], axis=1)
y = data['log_price']
```

```python
ct = make_column_transformer(
    (StandardScaler(), ['Mileage', 'EngineV']),
    (OneHotEncoder(), ['Brand', 'Body', 'Engine Type', 'Registration']),
    remainder='passthrough'
)
```

```python
x_scaled = ct.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42)
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
mae = mean_absolute_error(y_pred, y_test)
print("\nمتوسط الخطأ المطلق (MAE):", mae)
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.5, color='red')
plt.xlabel('القيم الحقيقية')
plt.ylabel('القيم المتوقعة')
plt.title('القيم الحقيقية مقابل التوقعات')
plt.show()
```

```python
with open('model.pkl', 'wb') as file:
    pickle.dump(reg, file)
```

```python
with open('column_transformer.pkl', 'wb') as file:
    pickle.dump(ct, file)
```
